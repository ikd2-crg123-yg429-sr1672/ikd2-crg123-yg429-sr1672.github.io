## Education Dataset

```python
import pandas as pd

education_df = pd.read_csv("../data/raw_data/Education2023.csv")

# Clean the column titles
education_df.columns = [
    col.strip()
    .lower()
    .replace(" ", "_")
    .replace(",", "")
    .replace('"', "")
    .replace("(", "")
    .replace(")", "")
    for col in education_df.columns
]

# Replace spaces with underscores and keep lowercase for 'area_name'
education_df["area_name"] = education_df["area_name"].str.replace(" ", "_")

for col in education_df.columns:
    if education_df[col].dtype == "object":
        education_df[col] = education_df[col].str.replace(",", "")

for col in education_df.columns:
    try:
        education_df[col] = pd.to_numeric(education_df[col], errors="ignore")
    except ValueError:
        continue  # This column contains non-numeric data, skip it


education_df.to_csv("../data/clean_data/education.csv", index=False)
education_df.head()
```

## Poverty Dataset

```python
poverty_df = pd.read_csv("../data/raw_data/Poverty2023.csv")

poverty_df.columns = [
    col.strip()
    .lower()
    .replace(" ", "_")
    .replace(",", "")
    .replace('"', "")
    .replace("(", "")
    .replace(")", "")
    for col in poverty_df.columns
]


poverty_df["area_name"] = poverty_df["area_name"].str.replace(" ", "_")

for col in poverty_df.columns:
    if poverty_df[col].dtype == "object":
        poverty_df[col] = poverty_df[col].str.replace(",", "")

for col in poverty_df.columns:
    try:
        poverty_df[col] = pd.to_numeric(poverty_df[col], errors="ignore")
    except ValueError:
        continue


poverty_df.to_csv("../data/clean_data/poverty.csv", index=False)
poverty_df.head()
```

## Unemployment Dataset

```python
unemployment_df = pd.read_csv("../data/raw_data/Unemployment2023.csv")

unemployment_df.columns = [
    col.strip()
    .lower()
    .replace(" ", "_")
    .replace(",", "")
    .replace('"', "")
    .replace("(", "")
    .replace(")", "")
    for col in unemployment_df.columns
]

unemployment_df["area_name"] = unemployment_df["area_name"].str.replace(
    r",\s*[A-Z]{2}$", "", regex=True
)
unemployment_df["area_name"] = unemployment_df["area_name"].str.replace(" ", "_")

for col in unemployment_df.columns:
    if unemployment_df[col].dtype == "object":
        unemployment_df[col] = unemployment_df[col].str.replace(",", "")

for col in unemployment_df.columns:
    try:
        unemployment_df[col] = pd.to_numeric(unemployment_df[col], errors="ignore")
    except ValueError:
        continue


unemployment_df.to_csv("../data/clean_data/unemployment.csv", index=False)
unemployment_df.head()
```
