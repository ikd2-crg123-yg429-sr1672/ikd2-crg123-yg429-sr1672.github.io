```python
import pandas as pd
import numpy as np
import altair as alt


df = pd.read_csv("../../data/clean_data/cleaned_addiction.csv")
df["Observation_12_month_ending_periods"] = pd.to_datetime(
    df["Observation_12_month_ending_periods"], format="%Y-%m-%d"
)

# group by 'Observation_12_month_ending_periods and then sum by Count_drug_overdose_deaths
df_grouped = (
    df.groupby("Observation_12_month_ending_periods")["Count_drug_overdose_deaths"]
    .sum()
    .reset_index()
)
```

```python
burgundy = "#8b0e4d"

line = (
    alt.Chart(df_grouped)
    .mark_line(
        color=burgundy, point=alt.OverlayMarkDef(color=burgundy, size=30), strokeWidth=3
    )
    .encode(
        x=alt.X("Observation_12_month_ending_periods:T", title="Date"),
        y=alt.Y("Count_drug_overdose_deaths:Q", title="Average Overdose Deaths"),
        tooltip=[
            alt.Tooltip("Observation_12_month_ending_periods:T", title="Date"),
            alt.Tooltip(
                "Count_drug_overdose_deaths:Q", title="Average Overdose Deaths"
            ),
        ],
    )
    .properties(title="12-Month Averaged Drug Overdose Deaths", width=450)
    .interactive()
)

line.save("../../outputs/smaller_addiction_lineplot.html")
line
```
